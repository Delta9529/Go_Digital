# Task 1 : Merging the data

```python
#Step 1
merge1 = pd.merge(emp,ins,on='AGENT_ID')
merge1.head()
```

```python
#Final Data
data = pd.merge(merge1,ven,how='left',on='VENDOR_ID')
data.head()
```

# Task 2 : Top 3 Insurance Type

```python
top_3 = data.groupby(['INSURANCE_TYPE'])['CLAIM_AMOUNT'].sum()
top_3.sort_values(ascending = False)
#Top 3 Insurance type are Life,Property and Health
```

# TASK 3 : Top 5 States

```python
risk_h = data[data['RISK_SEGMENTATION']=='H']
customer_state = risk_h['STATE_y'].value_counts()
print(customer_state.head())
#Top 5 states are CA,AZ,FL,TN,AR
```

# Task 4 : COLOCATION

```python
data['colocation'] = np.where(((data['STATE_x']==data['STATE_y']) & (data['STATE_y']==data['INCIDENT_STATE'])&(data['INCIDENT_STATE']==data['STATE_x'])),1,0)
print(data['colocation'].value_counts())
print(data['colocation'].mean())
# mean of the colocation column is 0.0044
```

## Task 5 : Data Entry Error

```python
data['AUTHORITY_CONTACTED'] = np.where(((data['POLICE_REPORT_AVAILABLE']==1) & (data['AUTHORITY_CONTACTED'] != 'Police')),'Police', data['AUTHORITY_CONTACTED'])
data[['AUTHORITY_CONTACTED','POLICE_REPORT_AVAILABLE']].head()
```

# Task 6 : Claim Deviation

```python
print(min(data['TXN_DATE_TIME']))
print(max(data['TXN_DATE_TIME']))
```

```python
avg_claim = data.loc[(data['TXN_DATE_TIME'] > '2021-05-30') & (data['TXN_DATE_TIME'] < '2021-06-29')]
avg_claim['CLAIM_AMOUNT'].mean()
#average claim amount
```

```python
data['CLAIM_DEVIATION'] = avg_claim['CLAIM_AMOUNT'].mean()/data['CLAIM_AMOUNT'] #Claim_deviation
data['claim_dev_status'] = np.where((data['CLAIM_DEVIATION']>0.5),1,0)
data[['CLAIM_DEVIATION','claim_dev_status']].head()
```

# Task 7 : Agents Insurance > 2


```python
agents2 = data.groupby(['AGENT_NAME','CLAIM_AMOUNT'])['INSURANCE_TYPE'].count().reset_index().sort_values(by = 'INSURANCE_TYPE',ascending = False).head()
agents2.head(15)
```

```python
top_agents = agents2[agents2['INSURANCE_TYPE']>2].sort_values(by = 'CLAIM_AMOUNT',ascending = False)
top_agents.head(25)
```
